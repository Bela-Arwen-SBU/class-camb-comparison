```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Properly expand the tilde to the full home path
file_path = os.path.expanduser('~/data/bella/class/output/explanatory01_cl.dat')

data = np.loadtxt(file_path)

# CLASS output columns (default): l, TT, EE, BB, TE
ell = data[:, 0]
cl_TT = data[:, 1]

# Convert to D_ell = l(l+1) C_ell / (2pi)
D_ell_TT = ell * (ell + 1) * cl_TT / (2 * np.pi)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(ell, D_ell_TT, label=r'$D_\ell^{TT}$')
plt.xlabel(r'Multipole $\ell$')
plt.ylabel(r'$D_\ell = \ell(\ell+1)C_\ell / 2\pi\ (\mu K^2)$')
plt.title('CMB TT Power Spectrum from CLASS')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

ell = data[:, 0]
TT = data[:, 1]
EE = data[:, 2]
BB = data[:, 3]
TE = data[:, 4]

plt.figure()
plt.plot(ell, ell*(ell+1)*TT/(2*np.pi), label='TT')
plt.plot(ell, ell*(ell+1)*EE/(2*np.pi), label='EE')
plt.plot(ell, ell*(ell+1)*BB/(2*np.pi), label='BB')
plt.plot(ell, ell*(ell+1)*TE/(2*np.pi), label='TE')
plt.xlabel(r'$\ell$')
plt.ylabel(r'$\ell(\ell+1)C_\ell/2\pi$ [$\mu K^2$]')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.title('CMB Power Spectra')
plt.show()
```

```python
pk_data = np.loadtxt(os.path.expanduser('~/data/bella/class/output/explanatory01_pk.dat'))
k = pk_data[:, 0]
Pk = pk_data[:, 1]

plt.figure()
plt.loglog(k, Pk)
plt.xlabel(r'$k$ [h/Mpc]')
plt.ylabel(r'$P(k)$ [Mpc$^3$]')
plt.title('Matter Power Spectrum')
plt.grid(True)
plt.show()
```

```python
from scipy.interpolate import interp1d

P_interp = interp1d(k, Pk, kind='cubic')
print(P_interp(0.1))  # Get P(k=0.1)
```
